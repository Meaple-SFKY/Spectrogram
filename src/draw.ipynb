```python
import torch
import matplotlib.pyplot as plt
```

```python
def bin_act(x):
	bin_wei = torch.sign(x).detach()
	out = torch.tanh(x)
	return bin_wei + out - out.detach()
```

```python
x = torch.arange(-2, 2, 0.001)
x.requires_grad = True
out = bin_act(x)
plt.plot(x.detach(), x.detach(), label='out-source')
plt.plot(x.detach(), out.detach(), label='out-forwarding')

out_sum = out.sum()
out_sum.backward()
plt.plot(x.detach(), x.grad, label='out-back-propagating')
plt.legend()
plt.show()
```
